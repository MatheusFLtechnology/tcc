```python
#Quantos projetos foram desenvolvidos por ano
pesquisa['inicio_execucao'] = pd.to_datetime(pesquisa['inicio_execucao'], errors='coerce')
pesquisa['fim_execucao'] = pd.to_datetime(pesquisa['fim_execucao'], errors='coerce')
#pesquisa['inicio_execucao'] = pesquisa['inicio_execucao'].astype('datetime64[ns]')
pesquisa.groupby(pesquisa.inicio_execucao.dt.year)['inicio_execucao'].count()
```

```python
#quantos projetos foram feitas por campus 
pesquisa.groupby('campus')['campus'].count()
```

```python
#5 campus com mais projetos de pesquisas 
#pesquisa.groupby('campus')['campus'].count()
pesquisa.groupby('campus')['campus'].sum().sort_values(ascending=False).head(6)
```

```python
#quantos projetos foram feitas por cordenador 
pesquisa.groupby('coordenador')['coordenador'].count().mean()
```

```python
#Quantos projetos foram desenvolvidos no campus canguaretama por ano?
pesquisa[pesquisa.campus == 'CANG'].groupby('campus')['campus'].count()
```
