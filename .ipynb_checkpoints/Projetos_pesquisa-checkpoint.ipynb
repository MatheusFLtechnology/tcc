# gerando os graficos

```python
import matplotlib as mpl
```

```python
import matplotlib.pyplot as plt
```

```python
# 5 áreas com a maior quantidade de projetos de pesquisa
pesquisa_area = pesquisa.groupby('area_conhecimento')['area_conhecimento'].count().sort_values(ascending=False).head(6)
x = pesquisa_area.index
y = pesquisa_area.values
plt.figure(figsize=(14,5))
plt.bar(x,y)
plt.show()
```

```python
# 5 áreas com a maior quantidade de projetos de pesquisa
pesquisa_area = pesquisa.groupby('area_conhecimento')['area_conhecimento'].count().sort_values(ascending=False).head(6)
x = pesquisa_area.index
y = pesquisa_area.values
plt.figure(figsize=(14,5))
plt.bar(x,y)
plt.show()
```

```python
# 5 campus com mais  projetos de pesquisa
pesquisa_campus = pesquisa.groupby('campus')['campus'].count().sort_values(ascending=False).head(5)
x = pesquisa_campus.index
y = pesquisa_campus.values
plt.figure(figsize=(9,5))
plt.bar(x,y)
plt.show()
```

# gerando graficos quntidades de projetos de pesquisa por pesquizadores

```python
# quantidades de projetos de pesquisa por pesquisador
#pesquisa.groupby('coordenador')['coordenador'].count().head(6)
pesquisa_pesquisador = pesquisa.groupby('coordenador')['coordenador'].count().sort_values(ascending=False).head(12)
x = pesquisa_pesquisador.index
y = pesquisa_pesquisador.values
plt.figure(figsize=(9,4))
plt.bar(x,y)
plt.xticks(rotation=50, ha='right')
plt.show()
```
